# 50 | Exercises - Bivariate Analysis within EDA

**Exploratory Data Analysis (EDA)** is the process of examining datasets to summarize datasets, identify patterns, detect anomalies, and explore relationships using visual and statistical methods. Types of EDA:

| Type | Variables Analyzed | Relationships Explored | Dependency Structure |
| --- | --- | --- | --- |
| **Univariate**   | One | None | None |
| **Bivariate**    | Two | Yes | One dependent variable |
| **Multivariate** | More than two | Yes | Multiple dependencies |

**Bivariate analysis** explores the relationship between two variables to identify patterns, correlations, and potential dependencies.

**Why use Bivariate Analysis in EDA?**

- Identify variables that are strong predictors for the model.
- Identify and drop highly correlated independent variables that may cause redundancy.
- Guidance to choose the right algorithm for a predictive model. For example, linear relationships work well with linear algorithms.
- Understanding the relationships between variables makes  it easier to interpret  predictive model outputs.
- Insights from bivariate analysis can inspire new features.
- In inferential statistics, it helps to form and test hypotheses.

## Correlation

**Correlation** is a statistical measure that describes the extent to which two variables are related: are changes in one variable associated with changes in another?

```python
import numpy as np
import matplotlib.pyplot as plt

# Create 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 4))

# Generate common x value
x = np.linspace(0, 10, 100)

# Positive or Direct Linear Relationship
y_positive = 2.5 * x + 3 + np.random.normal(0, 2, size=100)
axes[0].scatter(x, y_positive, alpha=0.7, color='b')
axes[0].set_xlabel("X values")
axes[0].set_ylabel("Y values")
axes[0].set_title("Positive or Direct Linear Relationship")

# Negative or Indirect Linear Relationship
y_negative = -2.5 * x + 3 + np.random.normal(0, 2, size=100)
axes[1].scatter(x, y_negative, alpha=0.7, color='r')
axes[1].set_xlabel("X values")
axes[1].set_ylabel("Y values")
axes[1].set_title("Negative or Indirect Linear Relationship")

# Negative or Indirect Linear Relationship
y_negative = -1 * x + 2 + np.random.normal(0, 2, size=100)
axes[2].scatter(x, y_negative, alpha=0.7, color='g')
axes[2].set_xlabel("X values")
axes[2].set_ylabel("Y values")
axes[2].set_title("Weaker Negative or Indirect Linear Relationship")

# Show plot with spacing adjustment to avoid overlap and cut offs
plt.tight_layout()
plt.show()
```

### Correlation Does Not Imply Causation

**Correlation** simply measures the strength and direction of a relationship between two variables. **Causation** requires a clear, direct link where changes in one variable directly result in changes in the other.

### Pearson Correlation

The **Pearson correlation coefficient (r)**, also known as **Pearsons r**, is a statistical measure that calculates the strength and direction of the linear relationship between two continuous numerical variables. It is appropriate when:

- both variables are **continuous**
- variable relationship is **linear**
- data is **normally distributed** or approximately symmetric
- **no extreme outliers** distorting correlation

**Strength and Direction of Pearson Correlation**

- **Postive** correlation: As one variable increases, the other also increases, e.g. more study, higher exam scores
- **Negative** correlation: As one variable increases, the other decreases, e.g. more tv, lower exam scores

| Positive | Negative | Correlation |
| --- | --- | --- |
| 0.00 to 0.09 | -0.00 to -0.09 | None |
| 0.10 to 0.39 | -0.10 to -0.39 | Weak |
| 0.40 to 0.69 | -0.40 to -0.69 | Moderate |
| 0.70 to 0.89 | -0.70 to -0.89 | Strong |
| 0.90 to 1.00 | -0.90 to -1.00 | Very Strong |

### Pearson Correlation with Python

```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load 'mpg' dataset from seaborn
mpg = sns.load_dataset('mpg')

# Step 3: Clean data
mpg_clean = mpg.dropna()

# Step 4: Calculate Pearson Correlation between 'mpg' and 'horsepower'
# Compare 'mpg' and 'horsepower' columns using the 'corr' method and specify 'pearson'
mpg_horsepower_corr = mpg_clean['mpg'].corr(mpg_clean['horsepower'], method='pearson')

# Step 5: Display the correlation result
print(f"The Pearson correlation coefficient is {mpg_horsepower_corr:.2f}.")
```

## Chi-Square Test

The **Chi-Square test** is a statistical method used to test the independence between two categorical variables.

| **Concept** | **Null Hypothesis (H)** | **Alternative Hypothesis (H)** |
| ---- | --- | --- |
| **Association** | Variables are _independent_ | Variables are _not independent_ |
| **Observed Differences** | Likely due to chance | Statistically significant |
| **p-value > 0.05** | Fail to reject H | Not enough evidence to support H |
| **p-value  0.05** | Reject H | Evidence supports H |
| **Conclusion** | No strong association detected | Meaningful association likely present |

## Chi-Square Test with Python

```python
# Step 1: Import libraries
import seaborn as sns
import pandas as pd
from scipy.stats import chi2_contingency

# Step 2: Load the 'tips' dataset from seaborn
tips = sns.load_dataset('tips')

# Step 3: Preview `tips` dataset
tips.head()
```

```python
# Step 4: Create a contingency table
contingency_table = pd.crosstab(tips['sex'], tips['smoker'])

# Step 5: Display contigency table
contingency_table
```

```python
# Step 6: Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Step 7: Print results
print("Chi-Square Statistic:", chi2)
print("P-Value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

# Step 8: Interpret results
if p < 0.05:
    print("Conclusion: There is a significant relationship between sex and smoker.")
else:
    print("Conclusion: There is no significant relationship between sex and smoker.")
```

## ANOVA (Analysis of Variance)

**ANOVA (Analysis of Variance)** is a statistical technique used to determine whether there are significant differences between the _means_ of three or more independent groups.

---

### Core Concepts

| Concept | Explanation |
| --- | --- |
| **Null Hypothesis (H)** | All group means are equal |
| **Alternative (H)** | At least one group mean is different |
| **Numerator** | _Between-Group Variance_: How much group means differ from the overall mean |
| **Denominator** | _Within-Group Variance_: Natural variation within each group |
| **F-Statistic** | Ratio of between-group to within-group variance: `F = Numerator / Denominator` |
| **P-Value** | Probability that observed differences are due to chance |
| **Degrees of freedom (df)** | Number of values that can vary independently when estimating statistical parameters, essential for determining the significance of the F-statistic |

---

### Types of ANOVA

| Type | Use Case |
| --- | --- |
| **One-Way ANOVA** | Compare means across one independent variable with 3+ groups (e.g. study methods on student performance) |
| **Two-Way ANOVA** | Compare means across two independent variables (e.g. study methods and environment on student performance) |
| **Repeated Measures** | Same subjects measured multiple times under different conditions (e.g. study methods on student performance over time) |

---

### Assumptions

Violating these assumptions can affect the reliability of ANOVA results. Violations can sometimes be addressed with alternative tests or corrections.

| Assumption | Meaning |
| --- | --- |
| **Normality** | Data in each group should be roughly normally distributed |
| **Equal Variances** | Group variances should be similar |
| **Independence** | Observations must be independent |

---

### Interpreting Results

| F-Statistic | With P-Value | Interpretation |
| --- | --- | --- |
| F  1 | p  0.05 | Fail to reject H  Group means likely not significantly different |
| F > 1 | p < 0.05 | Reject H  At least one group mean is significantly different |
| Very large F | p < 0.05 | Strong evidence of significant group differences |

**Example Interpretations**

| F-Statistic | With P-Value | Interpretation |
| --- | --- | --- |
| 1.2 | 0.30 | No significant difference  likely random chance |
| 8.5 | 0.001 | Significant difference  at least one group differs |

---

### Degrees of Freedom (df)

| Type |  Formula | Measure |
| --- | --- | --- |
| **Between Groups** | `df_between = k - 1` | Number of independent comparisons among group means |
| **Within Groups** | `df_within = N - k` | Variability within groups |
| **Total** | `df_total = N - 1` | Sum of these components |

## ANOVA with Python

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Step 2: Generate sample data
np.random.seed(42)

diet_A = np.random.normal(loc=5, scale=1.5, size=30)  # Mean weight loss ~5 kg
diet_B = np.random.normal(loc=6, scale=1.2, size=30)  # Mean weight loss ~6 kg
diet_C = np.random.normal(loc=7, scale=1.3, size=30)  # Mean weight loss ~7 kg

# Step 3: Combine into a DataFrame
data = pd.DataFrame({
    'weight_loss': np.concatenate([diet_A, diet_B, diet_C]),
    'diet_type': ['A'] * 30 + ['B'] * 30 + ['C'] * 30
})

# Step 4: Preview first few rows
print(data)
```

```python
# Step 5: One-Way ANOVA using statsmodels
model = ols('weight_loss ~ C(diet_type)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Step 6: Display the ANOVA table
print(anova_table)
```

### Interpreting ANOVA table

| Metric | Meaning |
| --- | --- |
| **C(diet_type)** | Between-Group Variation |
| **Residual** | Within-Group Variation |
| **Sum Sq** | Measures how much group means deviate from overall mean  higher value indicates greater differences |
| **df** | Degrees of Freedom |
| **F** | F-statistic |
| **PR(>F)** | P-value |


**Large F-statistic (25.12)** and **Small P-Value (0.00000000245)** indicates that observed differences are _statistically significant_. This _rejects_ the Null Hypothesis (H) and conclude that at least one diet leads to significantly different weight loss outcomes.

## Exercise 5

With the seaborn `titanic` dataset, determine whether there is a statistically significant relationship between survival status (`survived`) and passenger class (`pclass`).

```python
import pandas as pd
import seaborn as sns
from scipy.stats import chi2_contingency

titanic = sns.load_dataset('titanic')

contingency_table = pd.crosstab(titanic['survived'], titanic['pclass'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2)
print("P-Value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected, "\n")

if p < 0.05:
    print(" Conclusion: There is a significant relationship between survival and passenger class.")
else:
    print(" Conclusion: There is no significant relationship between survival and passenger class.")
```
