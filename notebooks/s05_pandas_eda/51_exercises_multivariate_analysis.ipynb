## Scatter Plot with Color Distinction

```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load 'iris' dataset from seaborn
iris = sns.load_dataset('iris')

# Step 3: Vew unique species
iris['species'].unique()
```

```python
# Step 4: Set the figure size
plt.figure(figsize=(8, 6))

# Step 5: Customize the Scatter Plot
# Select data, x-axis, y-axis and customizations
# Set hue to species and assign each specie a color
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    hue='species',
    palette={
        'setosa': 'blue',
        'versicolor': 'red',
        'virginica': 'green'
    }
)

# Step 6: Add title, labels and grid for readability
plt.title('Scatter Plot of Sepal Length vs Sepal Width with Species Color Distinction')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Step 7: Display plot
plt.tight_layout()
plt.show()
```

## Scatter Plot with Color and Size Distinction

```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load 'iris' dataset from seaborn
iris = sns.load_dataset('iris')

# Step 3: Vew unique species
iris['species'].unique()
```

```python
# Step 4: Set the figure size
plt.figure(figsize=(8, 6))

# Step 5: Customize the Scatter Plot
# Select data, x-axis, y-axis and customizations
# Set hue to species and assign each specie a color
# Set marker size to petal_length with min and max size
# Set marker transparency level to 0.7
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    hue='species',
    palette={
        'setosa': 'blue',
        'versicolor': 'red',
        'virginica': 'green'
        },
    size='petal_length',
    sizes=(20, 200),
    alpha=0.7
    )

# Step 6: Add title, labels and grid for readability
plt.title('Scatter Plot of Sepal Length vs Sepal Width with Species Color Distinction')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Step 7: Display plot and move legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

## Multivariate Line Plot

```python
# Step 1: Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Generate time series dataset with 100 daily rates
dates = pd.date_range(start='2020-01-01', periods=100, freq='D')

# Step 3: Create a DataFrame with random data
# temperature with mean 20°C and standard deviation 5
# humidity with mean 60% and standard deviation 10
# pressure with mean 1013 hPa and standard deviation 8
data = pd.DataFrame({
    'date': dates,
    'temperature': np.random.normal(loc=20, scale=5, size=100),
    'humidity': np.random.normal(loc=60, scale=10, size=100),
    'pressure': np.random.normal(loc=1013, scale=8, size=100)
    })

# Step 4: Set the figure size
plt.figure(figsize=(8, 6))

# Step 5: Creating a Multivariate Line Plot
# Assign x-axis, y-axis and label for legend for each line
plt.plot(data['date'], data['temperature'], label='Temperature (°C)')
plt.plot(data['date'], data['humidity'], label='Humidity (%)')
plt.plot(data['date'], data['pressure'], label='Pressure (hPa)')

# Step 6: Add title, labels and a legend for readability
plt.title('Multivariate Line Plot of Temperature, Humidity, and Pressure Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()

# Step : Display Plot
plt.show()
```

## Multivariate Box Plot

```python
# Step 1: Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # For 3D plotting (not used here)
import numpy as np

# Step 2: Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Step 3: Set the figure size
plt.figure(figsize=(8, 6))

# Step 4: Create a multivariate box plot
# Set hue to `sex`
sns.boxplot(
    data=penguins,
    x='species',
    y='flipper_length_mm',
    hue='sex',
    palette='Set2'
)

# Step 5: Add title and axis labels
plt.title('Box Plot of Flipper Length Across Penguin Species by Sex')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')

# Step 6: Display plot
plt.show()
```

## Multivariate Bar Plot

```python
# Step 1: Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Step 3: Set the figure size
plt.figure(figsize=(10, 6))

# Step 4: Create a multivariate bar plot
# Set hue to `sex` and remove confidence intervals (ci) for clearer comparison
# `ci` has been deprecated, use `errorbar` instead
sns.barplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    hue='sex',
    palette='Set2',
    errorbar=None
)

# Step 5: Add title and axis labels
plt.title('Bar Plot of Average Body Mass Across Penguin Species by Sex')
plt.xlabel('Species')
plt.ylabel('Average Body Mass (g)')

# Step 6: Display plot
plt.show()
```

Create a scatter plot using the seaborn `penguins` dataset. Plot `flipper_length_mm` against `bill_length_mm`. Map `body_mass_g` to the size of each data point, and use `species` to determine point color.

## Exercise 2

Create a scatter plot using the breast cancer dataset from scikit-learn. Plot `mean radius` against `mean texture` mapping `diagnosis` as the color of each data point.

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load breast cancer dataset from scikit-learn
data = load_breast_cancer(as_frame=True)
df = data.frame

# Rename target column for clarity
df['diagnosis'] = df['target'].map({0: 'malignant', 1: 'benign'})

plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=df,
    x='mean radius',
    y='mean texture',
    hue='diagnosis',
    palette="coolwarm",
    alpha=0.9
    )
plt.xlabel("Mean Radius")
plt.ylabel("Mean Texture")
plt.title("Mean Radius vs Mean Texture with Diagnosis Color Encoding")
plt.tight_layout()
plt.show()
```
