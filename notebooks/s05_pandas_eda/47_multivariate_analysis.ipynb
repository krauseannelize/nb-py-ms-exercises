## Multivariate Scatter Plot

**Multivariate Scatter Plots** visualize relationships between multiple numerical variables within a dataset. By incorporating features such as color, size, and transparency, it provides an intuitive way to explore patterns, interactions, and distributions across different categories

```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Load 'mpg' dataset from seaborn
mpg = sns.load_dataset('mpg')

# Step 3: Preview data
mpg.head()
```

```python
# Step 4: Set the figure size
plt.figure(figsize=(8, 6))

# Step 5: Create the scatterplot
# Set color distinction based on car's origin
# Adjust marker size according to weight and set transparency
sns.scatterplot(
    data=mpg,
    x='horsepower',
    y='mpg',
    hue='origin',
    palette='Set1',
    size='weight',
    sizes=(10, 100),
    alpha=1
)

# Step 4: Add title and labels
plt.title('Multivariate Scatter Plot of MPG by Horsepower and Weight')
plt.xlabel('Horsepower')
plt.ylabel('Miles Per Gallon (MPG)')

# Step 5: Display plot
plt.show()
```

## Multivariate Line Plot

**Multivariate line plots** visualize trends and variations in numerical data over time or another continuous variable, incorporating multiple dimensions for deeper analysis.

```python
# Step 1: Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Generate time series dataset with 100 daily rates
dates = pd.date_range(start='2020-01-01', periods=100, freq='D')

# Step 3: Create a DataFrame with random data
# temperature with mean 20C and standard deviation 5
# humidity with mean 60% and standard deviation 10
# pressure with mean 1013 hPa and standard deviation 8
data = pd.DataFrame({
    'date': dates,
    'temperature': np.random.normal(loc=20, scale=5, size=100),
    'humidity': np.random.normal(loc=60, scale=10, size=100),
    'pressure': np.random.normal(loc=1013, scale=8, size=100)
})

# Step 4: Preview data
data.head()
```

```python
# Step 5: Set the figure size
plt.figure(figsize=(8, 6))

# Step 6: Create a multivariate line plot
# Assign x-axis, y-axis and label for legend for each line
plt.plot(data['date'], data['temperature'], label='Temperature (C)')
plt.plot(data['date'], data['humidity'], label='Humidity (%)')
plt.plot(data['date'], data['pressure'], label='Pressure (hPa)')

# Step 7: Add title, labels, and legend
plt.title('Multivariate Line Plot of Temperature, Humidity, and Pressure Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()

# Step 8: Display plot
plt.show()
```

## Multivariate Bar Plot

**Multivariate bar plots** visualize the comparison of categorical and numerical data across multiple groups while incorporating an additional categorical variable to reveal deeper insights. By utilizing color differentiation and grouping, it allows for an effective representation of how different categories compare, facilitating pattern recognition and trend analysis.

```python
# Step 1: Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # For 3D plotting (not used here)
import numpy as np

# Step 2: Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Step 3: Preview data
penguins.head()
```

```python
# Step 4: Set the figure size
plt.figure(figsize=(8, 6))

# Step 5: Create a multivariate bar plot
# Set hue to `sex` and remove confidence intervals (ci) for clearer comparison
# `ci` has been deprecated, use `errorbar` instead
sns.barplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    hue='sex',
    palette='Set2',
    errorbar=None
)

# Step 6: Add title and axis labels
plt.title('Bar Plot of Average Body Mass Across Penguin Species by Sex')
plt.xlabel('Species')
plt.ylabel('Average Body Mass (g)')

# Step 7: Display plot
plt.show()
```

## Multivariate Box Plot

**Multivariate box plots** visualizations analyze the distribution and spread of numerical data across multiple categories while incorporating an additional categorical variable for deeper insights. By displaying quartiles, medians, and potential outliers, it provides a clear understanding of variability and central tendencies within the dataset.

```python
# Step 1: Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Step 3: Set the figure size
plt.figure(figsize=(8, 6))

# Step 4: Create a multivariate box plot
# Set hue to `sex` to explore how flipper length varies by sex within species
sns.boxplot(
    data=penguins,
    x='species',
    y='flipper_length_mm',
    hue='sex',
    palette='Set1'
)

# Step 5: Add title and axis labels
plt.title('Box Plot of Flipper Length Across Penguin Species by Sex')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')

# Step 6: Display the plot
plt.show()
```

## Multivariate Violin Plot

**Multivariate violin plots** combines aspects of both _box plots_ and _density plots_ to display the distribution of numerical data across multiple categories while incorporating an additional categorical variable for deeper analysis. This method allows for an intuitive representation of data spread, density, and central tendencies across multiple groups simultaneously. This kind of plots are particularly useful in identifying the shape of distributions, detecting multimodal patterns, and spotting outliers across different categories.

```python
# Step 1: Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Step 3: Set the figure size
plt.figure(figsize=(8, 6))

# Step 4: Create a multivariate violin plot
# Set hue to `sex` and split violin plot by hue variable
sns.violinplot(
    data=penguins,
    x='species',
    y='flipper_length_mm',
    hue='sex',
    split=True,
    palette='Paired'
)

# Step 5: Add title and axis labels
plt.title('Violin Plot of Flipper Length Across Penguin Species by Sex')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')

# Step 6: Display plot
plt.show()
```
