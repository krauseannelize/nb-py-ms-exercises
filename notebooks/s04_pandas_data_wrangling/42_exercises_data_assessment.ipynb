## Python Methods for DataFrame Assessment

### Checking the First Rows

The `.head()` method enables you to view the first 5 rows of a DataFrame if no argument is passed. You can also specify the number of rows to return by passing an integer as an argument.

````python
dirty_df.head(3)
````

### Checking the Shape of the DataFrame

The `.shape` attribute gives you the dimensions of the DataFrame as a tuple, showing the number of rows and columns.

````python
dirty_df.shape
````

### Checking the Data Type of Each Column

The `.dtypes` attribute returned the data types of the DataFrame.

````python
dirty_df.dtypes
````

### Identifying Missing Values

The `.isnull()` method flags missing entries as `True`. Chaining `.sum()` counts the number of missing values per column.

````python
dirty_df.isnull().sum()
````

### Checking Duplicate Rows

The `.duplicated()` method flags duplicated rows as `True`. Chaining `.sum()` returns the total number of duplicated rows in the DataFrame.

````python
dirty_df.duplicated().sum()
````

### Checking Unique Values

The `.unique()` method returns an array of unique values in a column. This can help identify inconsistencies, unexpected entries, or data formatting issues.

````python
dirty_df['Remarks'].unique()
````

### Assessment on Iris Dataset

````python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

# 1. Check the first few rows
print(df_iris.head())

# 2. Check the shape of the DataFrame
print('\nShape of the Iris dataset:')
print(df_iris.shape)

# 3. Check the data types of each column
print('\nData types of each column in Iris:')
print(df_iris.dtypes)

# 4. Check the number of missing values in the Iris dataset
print('\nMissing values in the Iris dataset:')
print(df_iris.isnull().sum())

# 5. Check how many duplicated rows are in the Iris dataset
print('\nDuplicated rows in the Iris dataset:')
print(df_iris.duplicated().sum())

# 6. Check unique values in the target column
print('\nUnique values in the target column:')
print(df_iris['target'].unique())
````
